```python
df1=df.copy()
```

```python
postal_to_county = df1.dropna(subset=['county']).set_index('postal_code')['county'].to_dict()

# Fill NaN values in 'county' based on postal_code
df1['county'] = df1.apply(lambda row: postal_to_county.get(row['postal_code'], row['county']) if pd.notna(row['postal_code']) else row['county'], axis=1)
```

```python
df1=df1.dropna(subset=["county"])
```

```python
county_to_city = df1.dropna(subset=['city']).set_index('county')['city'].to_dict()

# Fill NaN values in 'county' based on postal_code
df1['city'] = df1.apply(lambda row: county_to_city.get(row['county'], row['city']) if pd.notna(row['county']) else row['city'], axis=1)
```

```python
county_to_postal = df1.dropna(subset=['postal_code']).set_index('county')['postal_code'].to_dict()

# Fill NaN values in 'county' based on postal_code
df1['postal_code'] = df1.apply(lambda row: county_to_postal.get(row['county'], row['postal_code']) if pd.notna(row['county']) else row['postal_code'], axis=1)
```

```python
# Filling nan values for year_built by taking average of years for a specific postalcode
df1['year_built'].fillna(df1.groupby('postal_code')['year_built'].transform('mean'), inplace=True)
```

```python
df1["year_built"] = df1["year_built"].astype(int)
df1["year_built"]
```

```python
df1=df1.dropna(subset=["price"])
```

```python
df1['sqft'].fillna(df1.groupby(['postal_code', 'price'])['sqft'].transform('mean'), inplace=True)
```

```python
df1['sqft'].fillna(df1.groupby(['baths', 'price'])['sqft'].transform('mean'), inplace=True)
```

```python
df1['sqft'].fillna(df1.groupby( 'postal_code')['sqft'].transform('mean'), inplace=True)
```

```python
df1['beds']=df1['beds'].fillna(0).astype(int)
```

```python
df_grouped = df1.groupby(['sqft', 'price'])

  # Calculate the average (integer) for beds in each group
average_beds = df_grouped['beds'].transform(lambda x: int(x.mean()))

  # Replace 0 with the corresponding average in the 'beds' column
df1['beds'] = df1['beds'].where(df1['beds'] != 0, average_beds)
```

```python
df1['baths']=df1['baths'].fillna(0).astype(int)
```

```python
df_grouped = df1.groupby(['sqft', 'price','beds'])

  # Calculate the average (integer) for baths in each group
average_baths = df_grouped['baths'].transform(lambda x: int(x.mean()))

  # Replace 0 with the corresponding average in the 'baths' column
df1['baths'] = df1['baths'].where(df1['baths'] != 0, average_baths)
```
